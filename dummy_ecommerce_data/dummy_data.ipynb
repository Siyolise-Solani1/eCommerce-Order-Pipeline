```python
import pandas as pd 
from datetime import datetime,timedelta
import random
```

```python
# Generate Orders

orders = pd.DataFrame({
    'order_id': [f'55{i:03d}' for i in range(1,21)],
    'customer_id': [f'C00{random.randint(1,5)}' for _ in range(20)],
    'order_date': pd.date_range(start='2025-06-01',periods=20,freq='D')
})

orders.to_csv('orders.csv',index=False)
```

```python
# Generate Customers

Customers = pd.DataFrame({
    'customer_id': [f'C00{i}' for i in range(1,6)],
    'name': ['Alice','Amanda','Charlie','Alex','Kolisi'],
    'email': [f'user{i}@example.com' for i in range(1,6)]
})

Customers.to_csv('customers.csv',index=False)
```

```python
# Generate Shipments

statuses = ['Shipped','Delivered'] #'Placed','Cancelled'
shipments = pd.DataFrame({
    'shipment_id': [f'SHP-0{i}' for i in range(1,21)],
    'order_id': orders['order_id'],
    'status': [random.choice(statuses) for _ in range(20)],
    'shipped_at': pd.date_range(start='2025-06-02', periods=20, freq='D'),
    'delivered_at': pd.date_range(start='2025-06-03', periods=20, freq='D')

})
shipments.to_csv('shipments.csv', index=False)
```
